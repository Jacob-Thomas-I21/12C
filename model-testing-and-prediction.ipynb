```python
import pandas as pd
import numpy as np
import joblib
import tensorflow as tf
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import matplotlib.pyplot as plt
import seaborn as sns
from typing import Dict, List
import warnings
import os
warnings.filterwarnings('ignore')

print("TEST SUITE")
print("=" * 60)

class SimpleModelSelector:
   def __init__(self, models_path="/kaggle/input/models"):
       self.models_path = models_path
       self.models = {}
       self.feature_columns = [
           'primary_energy_consumption', 'energy_per_capita', 'oil_co2',
           'coal_co2', 'gdp', 'cement_co2', 'gas_co2', 'co2_per_unit_energy',
           'population', 'energy_per_gdp', 'co2_per_capita', 'methane',
           'nitrous_oxide', 'cumulative_co2'
       ]
       self.load_models()

   def load_models(self):
       model_files = {
           'xgboost': 'xgboost_model.pkl',
           'catboost': 'catboost_model.cbm',
           'random_forest': 'random_forest_tuned_model.pkl',
           'mlp_scaled': 'mlp_model_scaled.h5',
           'mlp_ae': 'mlp_model_ae.h5',
           'gmlp_ae': 'gmlp_model_ae.h5',
           'tabtransformer': 'tabtransformer_model_scaled.h5',
           'voting_ensemble': 'voting_ensemble_model.pkl',
           'stacking_ensemble': 'stacking_ensemble_model.pkl',
           'wide_deep': 'wide_deep_model.h5',
           'bayesian_mlp': 'bayesian_mlp_model_ae.h5',
           'ae_encoder': 'ae_encoder.h5',
           'ae_best': 'ae_best_model.h5'
       }

       print("Looking for models in Kaggle environment...")
       print(f"Base path: {self.models_path}")

       for model_name, filename in model_files.items():
           filepath = os.path.join(self.models_path, filename)
           if not os.path.exists(filepath):
               print(f"File not found: {filepath}")
               continue

           try:
               if filename.endswith('.cbm'):
                   from catboost import CatBoostRegressor
                   model = CatBoostRegressor()
                   model.load_model(filepath)
                   self.models[model_name] = model
                   print(f"Loaded CatBoost model: {model_name}")

               elif filename.endswith('.h5'):
                   if model_name == 'tabtransformer':
                       def expand_dims_layer(x):
                           return tf.expand_dims(x, axis=1)
                       custom_objects = {'expand_dims_layer': expand_dims_layer}
                       model = tf.keras.models.load_model(filepath, custom_objects=custom_objects, compile=False)
                       self.models[model_name] = model
                       print(f"Loaded Keras model with custom_objects: {model_name}")
                   else:
                       model = tf.keras.models.load_model(filepath, compile=False)
                       self.models[model_name] = model
                       print(f"Loaded Keras model: {model_name}")

               elif filename.endswith('.pkl'):
                   model = joblib.load(filepath)
                   self.models[model_name] = model
                   print(f"Loaded sklearn-compatible model: {model_name}")

               else:
                   print(f"Unrecognized extension for {filename}, skipping {model_name}")

           except Exception as e:
               print(f"Could not load {model_name} from {filepath}: {e}")

       if len(self.models) == 0:
           raise FileNotFoundError(
               "No models loaded! Please place your .pkl, .h5, or .cbm files in /kaggle/input/models/"
           )
       else:
           print(f"Total models loaded: {len(self.models)}")
           print(f"Successfully loaded: {list(self.models.keys())}")

   def select_best_model(self, country: str, year: int, gdp: float) -> str:
       if not self.models:
           return None

       if gdp > 50000:
           preferred = ['catboost', 'xgboost', 'tabtransformer', 'voting_ensemble']
           reason = "High-GDP countries: sensitive models for complex economies"

       elif gdp > 15000:
           preferred = ['catboost', 'tabtransformer', 'mlp_scaled', 'xgboost']
           reason = "Medium-GDP countries: CatBoost and TabTransformer handle emerging markets well"

       else:
           preferred = ['catboost', 'xgboost', 'random_forest']
           reason = "Low-GDP countries: CatBoost for better sensitivity"

       for model_name in preferred:
           if model_name in self.models:
               print(f"Selected: {model_name}")
               print(f"Reason: {reason}")
               return model_name

       fallback = list(self.models.keys())[0]
       print(f"Fallback: {fallback} (first available)")
       return fallback

   def predict_co2(self, model_name: str, features: np.ndarray) -> float:
       if model_name not in self.models:
           return None

       model = self.models[model_name]

       try:
           if hasattr(model, 'predict'):
               if isinstance(model, tf.keras.Model):
                   prediction = model.predict(features, verbose=0)[0]
                   if isinstance(prediction, np.ndarray):
                       prediction = prediction[0]
               else:
                   prediction = model.predict(features)[0]
           else:
               prediction = 1000.0

           return float(prediction)
       except Exception as e:
           print(f"Prediction failed for {model_name}: {e}")
           return None


def run_simple_analysis(country: str, year: int, baseline_features: Dict) -> Optional[Dict]:
   print(f"\nANALYZING: {country} ({year})")
   print("-" * 40)

   selector = SimpleModelSelector()

   gdp = baseline_features.get('gdp', 30000)
   best_model = selector.select_best_model(country, year, gdp)

   feature_array = np.array([baseline_features.get(col, 0) for col in selector.feature_columns]).reshape(1, -1)
   baseline_co2 = selector.predict_co2(best_model, feature_array)

   if baseline_co2 is None:
       print("Could not get baseline prediction")
       return None

   print(f"Baseline CO2: {baseline_co2:.2f}")

   scenarios = [
       {
           'name': 'Green Energy Transition',
           'changes': {'coal_co2': 0.1, 'oil_co2': 0.4, 'co2_per_unit_energy': 0.5}
       },
       {
           'name': 'Industrial Efficiency',
           'changes': {'cement_co2': 0.3, 'energy_per_gdp': 0.6, 'co2_per_unit_energy': 0.7}
       },
       {
           'name': 'Economic Decoupling',
           'changes': {'gdp': 1.15, 'co2_per_capita': 0.7, 'energy_per_gdp': 0.6}
       },
       {
           'name': 'Comprehensive Action',
           'changes': {'coal_co2': 0.2, 'oil_co2': 0.5, 'gas_co2': 0.6, 'cement_co2': 0.4}
       }
   ]

   results = []
   for scenario in scenarios:
       modified_features = baseline_features.copy()
       for feature, multiplier in scenario['changes'].items():
           if feature in modified_features:
               modified_features[feature] *= multiplier

       modified_array = np.array([modified_features.get(col, 0) for col in selector.feature_columns]).reshape(1, -1)
       scenario_co2 = selector.predict_co2(best_model, modified_array)

       if scenario_co2 is not None:
           reduction_pct = ((baseline_co2 - scenario_co2) / baseline_co2) * 100
           results.append({
               'scenario': scenario['name'],
               'baseline_co2': baseline_co2,
               'predicted_co2': scenario_co2,
               'reduction_pct': reduction_pct,
               'changes': scenario['changes']
           })
           print(f"{scenario['name']}: {reduction_pct:+.1f}%")

   if results:
       create_simple_visualization(results, country, year, best_model)

   return {
       'country': country,
       'year': year,
       'model_used': best_model,
       'baseline_co2': baseline_co2,
       'scenarios': results
   }


def create_simple_visualization(results: List[Dict], country: str, year: int, model_name: str):
   fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))
   scenarios = ['Baseline'] + [r['scenario'] for r in results]
   co2_values = [results[0]['baseline_co2']] + [r['predicted_co2'] for r in results]
   colors = ['gray'] + ['green' if r['reduction_pct'] > 0 else 'red' for r in results]

   ax1.bar(scenarios, co2_values, color=colors, alpha=0.7)
   ax1.set_ylabel('CO₂ Emissions')
   ax1.set_title(f'CO₂ Levels – {country} ({year})')
   ax1.tick_params(axis='x', rotation=45)
   for i, result in enumerate(results, 1):
       ax1.text(i, result['predicted_co2'] + max(co2_values) * 0.01,
                f'{result["reduction_pct"]:+.1f}%', ha='center', fontweight='bold')

   scenario_names = [r['scenario'] for r in results]
   reductions = [r['reduction_pct'] for r in results]
   colors2 = ['darkgreen' if r > 20 else 'green' if r > 10 else 'orange' if r > 0 else 'red' for r in reductions]

   ax2.bar(scenario_names, reductions, color=colors2, alpha=0.8)
   ax2.set_ylabel('CO₂ Reduction (%)')
   ax2.set_title(f'Reduction Potential – Model: {model_name}')
   ax2.tick_params(axis='x', rotation=45)
   ax2.axhline(0, color='black', linestyle='-', alpha=0.5)

   plt.tight_layout()
   plt.show()

   print(f"\nANALYSIS SUMMARY:")
   print(f"Country: {country} ({year})")
   print(f"Model Used: {model_name}")
   print(f"Baseline CO₂: {results[0]['baseline_co2']:.2f}")

   best_scenario = max(results, key=lambda x: x['reduction_pct'])
   print(f"Best Scenario: {best_scenario['scenario']} ({best_scenario['reduction_pct']:.1f}% reduction)")


def convert_company_to_national_scale(company_data: Dict) -> Dict:
   scaling_factors = {
       'manufacturing': {'energy_multiplier': 1000, 'emissions_multiplier': 500},
       'tech': {'energy_multiplier': 2000, 'emissions_multiplier': 800},
       'retail': {'energy_multiplier': 1500, 'emissions_multiplier': 600},
       'energy': {'energy_multiplier': 800, 'emissions_multiplier': 400},
       'healthcare': {'energy_multiplier': 1200, 'emissions_multiplier': 700},
       'finance': {'energy_multiplier': 3000, 'emissions_multiplier': 1000}
   }
   industry = company_data.get('industry', 'manufacturing')
   factors = scaling_factors.get(industry, scaling_factors['manufacturing'])

   employees = company_data.get('employees', 100)
   revenue = company_data.get('annual_revenue', 10_000_000)

   converted = {
       'gdp': revenue / 1000,
       'population': employees * 1000,
       'primary_energy_consumption': company_data.get('electricity_kwh', 500_000) / factors['energy_multiplier'],
       'energy_per_capita': company_data.get('electricity_kwh', 500_000) / (employees * 10),
       'energy_per_gdp': (company_data.get('electricity_kwh', 500_000) / factors['energy_multiplier']) / (revenue / 1_000_000),
       'oil_co2': company_data.get('fuel_oil_emissions', company_data.get('scope1_emissions', 1000) * 0.3) / factors['emissions_multiplier'],
       'coal_co2': company_data.get('coal_emissions', 0) / factors['emissions_multiplier'],
       'gas_co2': company_data.get('natural_gas_emissions', company_data.get('scope1_emissions', 1000) * 0.4) / factors['emissions_multiplier'],
       'cement_co2': company_data.get('industrial_process_emissions', company_data.get('scope1_emissions', 1000) * 0.1) / factors['emissions_multiplier'],
       'co2_per_unit_energy': company_data.get('scope1_emissions', 1000) / company_data.get('electricity_kwh', 500_000) * 1000,
       'co2_per_capita': company_data.get('total_emissions', company_data.get('scope1_emissions', 1000) + company_data.get('scope2_emissions', 500)) / employees,
       'methane': company_data.get('methane_emissions', company_data.get('scope1_emissions', 1000) * 0.05) / factors['emissions_multiplier'],
       'nitrous_oxide': company_data.get('n2o_emissions', company_data.get('scope1_emissions', 1000) * 0.02) / factors['emissions_multiplier'],
       'cumulative_co2': (company_data.get('scope1_emissions', 1000) + company_data.get('scope2_emissions', 500)) * company_data.get('years_operating', 10) / factors['emissions_multiplier']
   }

   industry_defaults = {
       'manufacturing': {'coal_co2': 0.5, 'co2_per_unit_energy': 0.6},
       'tech': {'coal_co2': 0.01, 'co2_per_unit_energy': 0.3},
       'retail': {'coal_co2': 0.1, 'co2_per_unit_energy': 0.4},
       'energy': {'coal_co2': 0.8, 'co2_per_unit_energy': 0.7},
       'healthcare': {'coal_co2': 0.05, 'co2_per_unit_energy': 0.45},
       'finance': {'coal_co2': 0.02, 'co2_per_unit_energy': 0.25}
   }
   defaults = industry_defaults.get(industry, industry_defaults['manufacturing'])

   for key, default_val in defaults.items():
       if converted.get(key, 0) == 0:
           converted[key] = default_val

   return converted


def test_developed_country() -> Optional[Dict]:
   baseline_features = {
       'gdp': 55000,
       'population': 80_000_000,
       'primary_energy_consumption': 12000,
       'energy_per_capita': 150,
       'oil_co2': 2500,
       'coal_co2': 1800,
       'gas_co2': 1200,
       'cement_co2': 300,
       'co2_per_unit_energy': 0.45,
       'energy_per_gdp': 0.25,
       'co2_per_capita': 9.5,
       'methane': 180,
       'nitrous_oxide': 70,
       'cumulative_co2': 45000
   }
   return run_simple_analysis("Germany", 2024, baseline_features)


def test_emerging_economy() -> Optional[Dict]:
   baseline_features = {
       'gdp': 12000,
       'population': 300_000_000,
       'primary_energy_consumption': 25000,
       'energy_per_capita': 80,
       'oil_co2': 3500,
       'coal_co2': 4500,
       'gas_co2': 800,
       'cement_co2': 800,
       'co2_per_unit_energy': 0.65,
       'energy_per_gdp': 0.4,
       'co2_per_capita': 4.2,
       'methane': 350,
       'nitrous_oxide': 120,
       'cumulative_co2': 25000
   }
   return run_simple_analysis("India", 2024, baseline_features)


def test_developing_country() -> Optional[Dict]:
   baseline_features = {
       'gdp': 3500,
       'population': 50_000_000,
       'primary_energy_consumption': 2500,
       'energy_per_capita': 45,
       'oil_co2': 800,
       'coal_co2': 400,
       'gas_co2': 200,
       'cement_co2': 150,
       'co2_per_unit_energy': 0.55,
       'energy_per_gdp': 0.6,
       'co2_per_capita': 2.8,
       'methane': 80,
       'nitrous_oxide': 30,
       'cumulative_co2': 5000
   }
   return run_simple_analysis("Bangladesh", 2024, baseline_features)


def test_manufacturing_company() -> Optional[Dict]:
   company_data = {
       'company_name': 'GreenTech Manufacturing',
       'industry': 'manufacturing',
       'employees': 500,
       'annual_revenue': 50_000_000,
       'electricity_kwh': 2_000_000,
       'scope1_emissions': 5_000,
       'scope2_emissions': 3_000,
       'fuel_oil_emissions': 1_500,
       'natural_gas_emissions': 2_000,
       'industrial_process_emissions': 500,
       'years_operating': 15
   }
   print(f"MANUFACTURING COMPANY: {company_data['company_name']}")
   print(f"Industry: {company_data['industry']} | Employees: {company_data['employees']:,}")
   print(f"Revenue: ${company_data['annual_revenue']:,} | Scope1+2: {company_data['scope1_emissions'] + company_data['scope2_emissions']:,} tons CO₂")

   national_features = convert_company_to_national_scale(company_data)
   return run_simple_analysis(company_data['company_name'], 2024, national_features)


def test_tech_company() -> Optional[Dict]:
   company_data = {
       'company_name': 'CloudTech Solutions',
       'industry': 'tech',
       'employees': 200,
       'annual_revenue': 25_000_000,
       'electricity_kwh': 800_000,
       'scope1_emissions': 100,
       'scope2_emissions': 400,
       'data_center_kwh': 500_000,
       'office_kwh': 300_000,
       'years_operating': 8
   }
   print(f"TECH COMPANY: {company_data['company_name']}")
   print(f"Industry: {company_data['industry']} | Employees: {company_data['employees']:,}")
   print(f"Revenue: ${company_data['annual_revenue']:,} | Total Emissions: {company_data['scope1_emissions'] + company_data['scope2_emissions']:,} tons CO₂")

   national_features = convert_company_to_national_scale(company_data)
   return run_simple_analysis(company_data['company_name'], 2024, national_features)


def test_retail_company() -> Optional[Dict]:
   company_data = {
       'company_name': 'EcoMart Retail Chain',
       'industry': 'retail',
       'employees': 2_000,
       'annual_revenue': 100_000_000,
       'electricity_kwh': 5_000_000,
       'scope1_emissions': 2_000,
       'scope2_emissions': 2_500,
       'years_operating': 20
   }
   print(f"RETAIL COMPANY: {company_data['company_name']}")
   print(f"Industry: {company_data['industry']} | Employees: {company_data['employees']:,}")
   print(f"Revenue: ${company_data['annual_revenue']:,} | Stores: {company_data['employees'] // 10}")

   national_features = convert_company_to_national_scale(company_data)
   return run_simple_analysis(company_data['company_name'], 2024, national_features)


def test_custom_scenario(country_name: str, features_dict: Dict) -> Optional[Dict]:
   return run_simple_analysis(country_name, 2024, features_dict)


def test_custom_company(company_data: Dict) -> Optional[Dict]:
   print(f"CUSTOM COMPANY: {company_data.get('company_name', 'Unknown')}")
   print(f"Industry: {company_data.get('industry', 'N/A')} | Employees: {company_data.get('employees', 0):,}")

   national_features = convert_company_to_national_scale(company_data)
   return run_simple_analysis(company_data.get('company_name', 'Custom Company'), 2024, national_features)


def run_all_tests() -> List[Dict]:
   print("RUNNING COMPREHENSIVE MODEL SELECTION TESTS")
   print("=" * 60)

   print("\n" + "=" * 50)
   print("TEST 1: DEVELOPED COUNTRY")
   print("=" * 50)
   result1 = test_developed_country()

   print("\n" + "=" * 50)
   print("TEST 2: EMERGING ECONOMY")
   print("=" * 50)
   result2 = test_emerging_economy()

   print("\n" + "=" * 50)
   print("TEST 3: DEVELOPING COUNTRY")
   print("=" * 50)
   result3 = test_developing_country()

   print("\n" + "=" * 50)
   print("MODEL SELECTION SUMMARY")
   print("=" * 50)

   results = [r for r in [result1, result2, result3] if r is not None]
   models_used = {r['model_used'] for r in results}

   for result in results:
       best = max(result['scenarios'], key=lambda x: x['reduction_pct'])
       print(f"{result['country']}: {result['model_used']} -> Best: {best['scenario']}")

   print(f"\nModels tested: {models_used}")
   print("Country tests completed!")

   return results


def run_company_test_suite() -> List[Dict]:
   print("RUNNING COMPANY DATA TEST SUITE")
   print("=" * 60)

   company_tests = [
       ("Manufacturing", test_manufacturing_company),
       ("Technology", test_tech_company),
       ("Retail", test_retail_company)
   ]

   company_results = []

   for test_name, test_func in company_tests:
       print(f"\n" + "=" * 50)
       print(f"COMPANY TEST: {test_name.upper()}")
       print("=" * 50)

       try:
           result = test_func()
           if result:
               company_results.append(result)
               best = max(result['scenarios'], key=lambda x: x['reduction_pct'])
               print(f"\n{test_name} Summary:")
               print(f"   Model: {result['model_used']}")
               print(f"   Baseline CO₂: {result['baseline_co2']:.2f}")
               print(f"   Best Scenario: {best['scenario']} ({best['reduction_pct']:+.1f}%)")
       except Exception as e:
           print(f"{test_name} test failed: {e}")

   if company_results:
       print(f"\n" + "=" * 60)
       print("COMPANY COMPARISON SUMMARY")
       print("=" * 60)

       for result in company_results:
           best = max(result['scenarios'], key=lambda x: x['reduction_pct'])
           print(f"{result['country']:25} | Model: {result['model_used']:15} | Best: {best['reduction_pct']:+5.1f}% | {best['scenario']}")

   return company_results


def run_complete_test_suite() -> Dict:
   print("RUNNING COMPLETE TEST SUITE: COUNTRIES + COMPANIES")
   print("=" * 70)

   print("\nPART 1: COUNTRY ANALYSIS")
   country_results = run_all_tests()

   print("\nPART 2: COMPANY ANALYSIS")
   company_results = run_company_test_suite()

   print(f"\n" + "=" * 70)
   print("COMPLETE TEST SUITE SUMMARY")
   print("=" * 70)

   print(f"Countries tested: {len(country_results)}")
   print(f"Companies tested: {len(company_results)}")

   models_used = set()
   models_used.update(r['model_used'] for r in country_results)
   models_used.update(r['model_used'] for r in company_results)

   print(f"Total models tested: {len(models_used)}")
   print(f"Models used: {models_used}")

   if country_results and company_results:
       best_overall = None
       best_reduction = float('-inf')

       for result in country_results + company_results:
           if result and result['scenarios']:
               best = max(result['scenarios'], key=lambda x: x['reduction_pct'])
               if best['reduction_pct'] > best_reduction:
                   best_reduction = best['reduction_pct']
                   best_overall = (result['country'], best['scenario'], best_reduction)

       if best_overall:
           print(f"\nBEST PERFORMER: {best_overall[0]}")
           print(f"   Best Scenario: {best_overall[1]}")
           print(f"   CO₂ Reduction: {best_overall[2]:.1f}%")

   print("\nALL TESTS COMPLETED SUCCESSFULLY!")

   return {
       'country_results': country_results,
       'company_results': company_results,
       'total_tests': len(country_results) + len(company_results),
       'models_tested': list(models_used)
   }


def quick_test(country="Germany", gdp=50000) -> Optional[Dict]:
   baseline = {
       'gdp': gdp,
       'population': 80_000_000 if gdp > 40000 else 200_000_000,
       'primary_energy_consumption': 12000 if gdp > 40000 else 8000,
       'energy_per_capita': 120 if gdp > 40000 else 60,
       'oil_co2': 2000,
       'coal_co2': 1500 if gdp > 40000 else 3000,
       'gas_co2': 1000,
       'cement_co2': 400,
       'co2_per_unit_energy': 0.4 if gdp > 40000 else 0.6,
       'energy_per_gdp': 0.3 if gdp > 40000 else 0.5,
       'co2_per_capita': 8 if gdp > 40000 else 4,
       'methane': 150,
       'nitrous_oxide': 60,
       'cumulative_co2': 30000
   }
   return run_simple_analysis(country, 2024, baseline)


def compare_models_performance():
   selector = SimpleModelSelector()
   if len(selector.models) < 2:
       print("Need at least 2 models for comparison")
       return

   test_features = {
       'gdp': 45000,
       'population': 100_000_000,
       'primary_energy_consumption': 10000,
       'energy_per_capita': 100,
       'oil_co2': 2000,
       'coal_co2': 2500,
       'gas_co2': 1000,
       'cement_co2': 400,
       'co2_per_unit_energy': 0.5,
       'energy_per_gdp': 0.35,
       'co2_per_capita': 6.5,
       'methane': 200,
       'nitrous_oxide': 80,
       'cumulative_co2': 35000
   }

   feature_array = np.array([test_features.get(col, 0) for col in selector.feature_columns]).reshape(1, -1)

   print("MODEL PERFORMANCE COMPARISON")
   print("-" * 40)

   predictions = {}
   for model_name in selector.models:
       pred = selector.predict_co2(model_name, feature_array)
       if pred is not None:
           predictions[model_name] = pred
           print(f"{model_name}: {pred:.2f}")

   if predictions:
       avg_pred = np.mean(list(predictions.values()))
       print(f"\nAverage prediction: {avg_pred:.2f}")

       print("\nVariance from average:")
       for model, pred in predictions.items():
           diff = ((pred - avg_pred) / avg_pred) * 100
           print(f"{model}: {diff:+.1f}%")


run_complete_test_suite()
```
